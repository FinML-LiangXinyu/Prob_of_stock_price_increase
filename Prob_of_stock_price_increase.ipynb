```python
import numpy as np

# 对数收益率的均值和标准差
mu = 0.01
sigma = 0.3
# 0 时刻的股价
S_0 = 25
# 时间周期 T
T = 250

# 生成一个股价随机游走过程
x = np.random.normal(0, 1, T)
x[0] = 0
W_process = x.cumsum()

S_process = S_0 * np.exp((mu - sigma ** 2 / 2) * sigma * W_process)
```

```python
from scipy.stats import norm
# 对数正态分布假设下股价上涨的概率
Norm_Prob = [norm.cdf(mu / sigma* delta_t ** 0.5) for delta_t in range(1, len(S_process))]
```

```python
# 布朗运动假设下股价上涨的概率
Wiener_Prob = [norm.cdf((mu - 1 / 2 * sigma ** 2) / sigma * delta_t ** 0.5) for delta_t in range(1, len(S_process))]
```

```python
import matplotlib.pyplot as plt

# 绘制股价随时间趋势的上涨概率
plt.plot(Norm_Prob, label = "LogNorm")
# 绘制股价随时间趋势的上涨概率
plt.plot(Wiener_Prob, label = "Wiener Process")

plt.legend()
```
